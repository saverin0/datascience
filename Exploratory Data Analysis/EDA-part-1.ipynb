```python
#Missing value
dataset.isnull().sum()
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
#features with nan values
features_with_na
```

```python
dataset['Electrical'].isnull().mean()
```

```python
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean()*100,5),'% missing values')
```

```python
sns.boxplot(dataset['SalePrice'])
#so many outliers in box plot
```

```python
#Let's plot some diagrams 

for feature in features_with_na:
    data = dataset.copy()
    #Let's make a variable that indicate 1 if the observation was missing
    data[feature]=np.where(data[feature].isnull(),1,0)

    #Lets calculate the mean SalePrice where the inofrmation is mising or present
    data.groupby(feature)['SalePrice'].median().plot.bar() #to avoid outliers i used median
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section

### Numerical Variables

```python
dataset.info()
```

```python
#list to check numerical variables

dataset['SaleType'].dtypes!='O'
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
## Temporal Variable (Eg. Datetime Variables)
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
#Here we will compare the difference between all years features with Salesprice
data=dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature] #age of house

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### Observations:
Within all the the yearwise information, it is clearly visible that the the New houses are more costlier that the older one. looks like a pareto or power law distribution

```python
### numerical variables - 2 types
##1. Continuous variable and Discrete variabe

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<=25]
```

```python
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
## let's find the relationship between Discrete and Sales Price
data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## There is a relationship between Discrete features and SalePrice
```

```python
## let's find the relationship between Continuous and Sales Price
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature]
```

```python
print(len(continuous_feature))
```

```python
data = dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
### Lets analyze the continous values by creating histograms to understand the distribution
data=dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

### EDA PART 2

```python
## We will be using logarithmic transformation
data=dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
## Outliers
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
