```python
# Replace NaN with 0 in specific columns
mq[['qty', 'altQty', 'netWgt']] = mq[['qty', 'altQty', 'netWgt']].fillna(0)

# Updated function with debugging
def determine_effective_qty(row, gold_prices):
    # Extract the values for comparison
    values = [row['qty'], row['altQty'], row['netWgt']]
    print(f"Row values: {values}")  # Debug: print row values

    # Find the maximum value
    max_value = max(values)

    # Check if max_value is invalid (0)
    if max_value == 0:
        # Check if primaryValue is valid
        if not pd.isna(row['primaryValue']) and row['primaryValue'] > 0:
            # Fetch the gold price for the given year and month
            price = gold_prices.get((row['refYear'], row['refMonth']))
            print(f"Year: {row['refYear']}, Month: {row['refMonth']}, Price: {price}")  # Debug: gold price lookup
            if price:
                # Calculate effective_qty using primaryValue and gold price
                result = row['primaryValue'] / price
                print(f"Effective qty (primaryValue / price): {result}")  # Debug: calculated value
                return result
            else:
                print(f"Gold price not found for {(row['refYear'], row['refMonth'])}")  # Debug: missing price
                return 0  # Return 0 if price is missing
    print(f"Returning max_value: {max_value}")  # Debug: returning max_value
    return max_value  # Return max_value if it's valid
```

```python
# Apply the function to the DataFrame
mq['effective_qty'] = mq.apply(determine_effective_qty, axis=1, args=(gold_prices,))
```

```python
# Remove duplicates and exclude rows where `partnerDesc` is "World"
mq_cleaned = mq.drop_duplicates()
mq_cleaned = mq_cleaned[mq_cleaned['partnerDesc'] != 'World']
```

```python
# Group by refYear, reporterDesc, and partnerDesc
grouped = mq_cleaned.groupby(['refYear', 'reporterDesc', 'partnerDesc'])

# Function to check the condition for each group and sub-group based on refMonth
def check_scenarios(group):
    # Group by refMonth within each main group
    subgroups = group.groupby('refMonth')

    results = []
    for _, subgroup in subgroups:
        first_entry = subgroup['primaryValue'].iloc[0]
        sum_of_others = subgroup['primaryValue'].iloc[1:].sum()
        # Add a column indicating whether the condition holds
        subgroup['is_sum_valid'] = first_entry == sum_of_others
        results.append(subgroup)

    # Combine the results for all subgroups
    return pd.concat(results)

# Apply the function to each main group
validated_df = grouped.apply(check_scenarios).reset_index(drop=True)

# Remove the grouping columns manually if needed
validated_df = validated_df.loc[:, ~validated_df.columns.duplicated()]

# Check if any True is available in the is_sum_valid column
true_values = validated_df[validated_df['is_sum_valid'] == True]

#Display rows with True values
#print(true_values)

 # Display the DataFrame with the validation result
print(validated_df)
```

```python
import pandas as pd

# Function to apply the filtering logic per group
def resolve_conflicts(group):
    # Separate true and false rows
    true_rows = group[group['is_sum_valid'] == True]
    false_rows = group[group['is_sum_valid'] == False]

    # If no true rows or just one, no special logic needed
    if len(true_rows) <= 1:
        return pd.concat([true_rows, false_rows], ignore_index=True)

    # Multiple true rows: keep only the ones with the max primaryValue
    max_val = true_rows['primaryValue'].max()
    top_rows = true_rows[true_rows['primaryValue'] == max_val]

    # If multiple rows have the same max primaryValue, keep only the first one
    if len(top_rows) > 1:
        # Sort top_rows by some stable criterion (like original index order) to get the first occurrence
        top_rows = top_rows.sort_index()  # ensure we respect original order
        top_rows = top_rows.iloc[[0]]  # keep only the first

    # Combine the selected top row(s) with the unchanged false rows
    combined = pd.concat([top_rows, false_rows], ignore_index=True)
    return combined

# Apply the logic per group of refYear, refMonth, reporterDesc, partnerDesc
group_cols = ['refYear', 'refMonth', 'reporterDesc', 'partnerDesc']
result_df = validated_df.groupby(group_cols, group_keys=False).apply(resolve_conflicts).reset_index(drop=True)

print(result_df)
```

```python
duplicates = result_df.duplicated(subset=["refYear", "refMonth", "reporterDesc", "partnerDesc", "primaryValue"])

# Drop duplicates while keeping the first occurrence
df_no_duplicates = result_df.drop_duplicates(subset=["refYear", "refMonth", "reporterDesc", "partnerDesc", "primaryValue"], keep="first")

# Display results
print("Duplicates:\n", duplicates)
print("\nDataFrame after dropping duplicates:\n", df_no_duplicates)
```
