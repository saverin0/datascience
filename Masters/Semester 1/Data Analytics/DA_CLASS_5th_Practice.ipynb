```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python

# Sample data
data = {'hours_studied': [1, 2, 3, 4, 5],
        'exam_score': [65, 70, 75, 80, 85]}
df = pd.DataFrame(data)
```

```python

# Create a linear regression model
model = LinearRegression()
```

```python

# Fit the model to the data
model.fit(df[['hours_studied']], df['exam_score'])
```

```python
# Print the model coefficients
print(model.coef_)  # Slope
```

```python
print(model.intercept_)  # Y-intercept
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'hours_studied': [1, 2, 3, 4, 5, 6, 7, 8],
        'exam_score': [65, 70, 78, 82, 85, 90, 92, 95]}
df = pd.DataFrame(data)

# Create the regression plot
sns.regplot(x='hours_studied', y='exam_score', data=df)
plt.show()
```

```python
# Create the residual plot
sns.residplot(x='hours_studied', y='exam_score', data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate sample data with a linear relationship
np.random.seed(0)  # For reproducibility
x = np.random.rand(50) * 10  # Independent variable
y = 2 * x + 1 + np.random.randn(50) * 2  # Dependent variable with some noise

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Create the residual plot
sns.residplot(x='x', y='y', data=df)
plt.title('Residual Plot')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Residuals')
plt.show()
```

```python

# Now, let's introduce some non-linearity
y_nonlinear = x**2 + 1 + np.random.randn(50) * 2
df['y_nonlinear'] = y_nonlinear

# Create the residual plot for the non-linear data
sns.residplot(x='x', y='y_nonlinear', data=df)
plt.title('Residual Plot (Non-linear Data)')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Residuals')
plt.show()
```
