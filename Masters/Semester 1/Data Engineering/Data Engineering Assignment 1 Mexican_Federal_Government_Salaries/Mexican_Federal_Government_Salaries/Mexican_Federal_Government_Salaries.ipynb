```python
#montoneto is float64 data type we can use mean of this
```

```python
#calculate the average of the column
avg_montoneto = df['montoneto'].astype("float").mean(axis=0)
```

```python
print("Average of montoneto:",avg_montoneto)
```

### Replace "NaN" by mean value by the mean value in "montoneto" column

```python
import numpy as np
df['montoneto'].replace(np.nan,avg_montoneto,inplace = True)
```

```python
avg_montobruto = df['montobruto'].astype("float").mean(axis=0)
```

```python
print("Average of montobruto:",avg_montobruto)
```

### Replace "NaN" by mean value by the mean value in "montobruto" column

```python
df['montobruto'].replace(np.nan,avg_montobruto,inplace = True)
```

```python
columns_with_missing_values_with_dtype_object = ['nombre', 'denominacion', 'cargo', 'area']

# Filtering the rows where columns have the missing values
finding_common_rows_with_missing = df[df[columns_with_missing_values_with_dtype_object].isnull().any(axis=1)]
```
