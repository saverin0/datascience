```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for column in combine_df.columns:
    unique_values = combine_df[column].unique()
    print(f"Unique values in '{column}': {unique_values}")
```

```python
# Count how many columns have only one unique value
single_value_columns = combine_df.nunique() == 1
count_single_value_columns = single_value_columns.sum()

print(f"Number of columns with only a single unique value: {count_single_value_columns}")

# Optionally, get the names of these columns
columns_with_single_value = combine_df.columns[single_value_columns].tolist()
print(f"Columns with only a single unique value: {columns_with_single_value}")
```
