```python
# 1. Total FOB Value by Reporter Country
reporter_fob_value = df.groupby('reporterDesc')['fobvalue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
reporter_fob_value.plot(kind='bar', color='skyblue')
plt.title('Total FOB (Free On Board) Value by Reporter Country')
plt.xlabel('Reporter Country')
plt.ylabel('FOB (Free On Board) Value (in billions)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 2. Net Weight Distribution
plt.figure(figsize=(10, 6))
df['netWgt'].plot(kind='hist', bins=50, color='orange', edgecolor='black')
plt.title('Net Weight Distribution')
plt.xlabel('Net Weight (kg)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 3. Trade Flow by Commodity
commodity_fob_value = df.groupby('cmdDesc')['fobvalue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
commodity_fob_value.plot(kind='bar', color='green')
plt.title('FOB Value by Commodity (Top 10)')
plt.xlabel('Commodity Description')
plt.ylabel('FOB Value (in billions)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 4. Yearly Trends in FOB Value
yearly_fob_value = df.groupby('refYear')['fobvalue'].sum()
plt.figure(figsize=(10, 6))
yearly_fob_value.plot(kind='line', marker='o', linestyle='-', color='blue')
plt.title('Yearly Trends in FOB Value')
plt.xlabel('Year')
plt.ylabel('FOB Value (in billions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 8. Count of 0.0 and NaN values in 'qty' and 'altQty'
qty_zeros = (df['qty'] == 0).sum()
altQty_zeros = (df['altQty'] == 0).sum()
qty_nans = df['qty'].isna().sum()
altQty_nans = df['altQty'].isna().sum()

print(f"Count of 0.0 values in 'qty': {qty_zeros}")
print(f"Count of NaN values in 'qty': {qty_nans}")
print(f"Count of 0.0 values in 'altQty': {altQty_zeros}")
print(f"Count of NaN values in 'altQty': {altQty_nans}")
```

```python
# 9. Compare which column has values and update 'qty' if it has 0.0
# Create a mask where 'qty' is 0.0 and 'altQty' is not null
df.loc[(df['qty'] == 0) & (df['altQty'].notna()), 'qty'] = df['altQty']

print("Updated 'qty' values where it was 0.0 and 'altQty' had data:")
print(df[['refYear', 'reporterDesc', 'partnerDesc', 'qty', 'altQty', 'altQtyUnitAbbr']])
```

```python
# Convert 'altQty' values to kg if 'altQtyUnitAbbr' is 'g'
mask = df['altQtyUnitAbbr'] == 'g'
df.loc[mask, 'altQty'] = df.loc[mask, 'altQty'] / 1000

df['altQtyUnitAbbr'] = df['altQtyUnitAbbr'].replace('g', 'kg')
```

```python
# 10. Graphs showing quantity bought by countries from Sudan or Zambia
# Filter the data for Sudan or Zambia
sudan_zambia_data = df[(df['reporterDesc'] == 'Sudan') | (df['reporterDesc'] == 'Zambia')]

# Plot the quantity bought by partner countries from Sudan or Zambia
plt.figure(figsize=(14, 8))
sns.barplot(data=sudan_zambia_data, x='partnerDesc', y='qty', hue='reporterDesc', ci=None)
plt.title('Quantity Bought by Partner Countries from Sudan or Zambia')
plt.xlabel('Partner Country (Buying Country)')
plt.ylabel('Quantity')
plt.xticks(rotation=90)
plt.legend(title='Reporter Country (Selling Country)')
plt.tight_layout()
plt.show()
```

```python
# Plot the yearly quantity bought from Sudan or Zambia using a stacked bar chart
plt.figure(figsize=(14, 8))
sudan_qty_per_year = sudan_zambia_data[sudan_zambia_data['reporterDesc'] == 'Sudan'].groupby(['refYear', 'partnerDesc'])['qty'].sum().unstack()
zambia_qty_per_year = sudan_zambia_data[sudan_zambia_data['reporterDesc'] == 'Zambia'].groupby(['refYear', 'partnerDesc'])['qty'].sum().unstack()

sudan_qty_per_year.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='Blues', width=0.8)
zambia_qty_per_year.plot(kind='bar', stacked=True, ax=plt.gca(), cmap='Greens', width=0.5, alpha=0.7)

plt.title('Yearly Quantity Bought from Sudan or Zambia by Partner Countries (Stacked Bar Chart)')
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.legend(title='Partner Country (Buying Country)')
plt.tight_layout()
plt.show()
```

```python
# Create a heatmap to visualize the quantity bought from Sudan or Zambia by partner countries over the years
pivot_table = sudan_zambia_data.pivot_table(values='qty', index='partnerDesc', columns='refYear', aggfunc='sum', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt='.2f', cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Quantity Bought from Sudan or Zambia by Partner Countries Over the Years')
plt.xlabel('Year')
plt.ylabel('Partner Country (Buying Country)')
plt.tight_layout()
plt.show()
```

```python
# 12. Clustering Partner Countries by FOB Value
partner_fob_values = df.groupby('partnerDesc')['fobvalue'].sum().reset_index()

# Standardize the FOB values for clustering
scaler = StandardScaler()
partner_fob_values_scaled = scaler.fit_transform(partner_fob_values[['fobvalue']])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
partner_fob_values['Cluster'] = kmeans.fit_predict(partner_fob_values_scaled)

# Plot the clusters
plt.figure(figsize=(14, 8))
sns.barplot(data=partner_fob_values, x='partnerDesc', y='fobvalue', hue='Cluster', palette='viridis')
plt.title('Clustering of Partner Countries by FOB Value')
plt.xlabel('Partner Country (Buying Country)')
plt.ylabel('FOB Value (in billions)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
