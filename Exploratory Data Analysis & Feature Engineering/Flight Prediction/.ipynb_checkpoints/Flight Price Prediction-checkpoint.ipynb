### Flight Price Prediction(EDA + Feature Engineering)

```python
## Dataset: https://github.com/krishnaik06/5-Days-Live-EDA-and-Feature-Engineering/tree/main/Flight%20Prediction
```

```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_excel('Data_Train.xlsx')
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset['Date']=dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month']=dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year']=dataset['Date_of_Journey'].str.split('/').str[2]
```

```python
dataset.head(2)
```

```python
dataset['Date'] = dataset['Date_of_Journey'].apply(lambda x:x.split("/")[0])
dataset['Month'] = dataset['Date_of_Journey'].apply(lambda x:x.split("/")[1])
dataset['Year'] = dataset['Date_of_Journey'].apply(lambda x:x.split("/")[2])
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset['Date']=dataset['Date'].astype(int)
dataset['Month']=dataset['Month'].astype(int)
dataset['Year']=dataset['Year'].astype(int)
```

```python
dataset.info()
```

```python
dataset.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
dataset.head(2)
```

```python
#dataset['Arrival_Time']=dataset['Arrival_Time'].str.split(' ').str[0]
dataset['Arrival_Time']=dataset['Arrival_Time'].apply(lambda x : x.split(' ')[0])
dataset['Arrival_Hour']=dataset['Arrival_Time'].str.split(':').str[0]
dataset['Arrival_Minute']=dataset['Arrival_Time'].str.split(':').str[1]
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
dataset['Arrival_Hour']=dataset['Arrival_Hour'].astype(int)
dataset['Arrival_Minute']=dataset['Arrival_Minute'].astype(int)
```

```python
dataset.info()
```

```python
dataset.drop('Arrival_Time',axis=1,inplace=True)
```

```python
dataset.head(2)
```

```python
#dataset['Arrival_Time']=dataset['Arrival_Time'].str.split(' ').str[0]
dataset['Dep_Hour']=dataset['Dep_Time'].str.split(':').str[0]
dataset['Dep_Minute']=dataset['Dep_Time'].str.split(':').str[1]
dataset['Dep_Hour']=dataset['Dep_Hour'].astype(int)
dataset['Dep_Minute']=dataset['Dep_Minute'].astype(int)
```

```python
dataset.head(2)
```

```python
dataset.drop('Dep_Time',axis=1,inplace=True)
```

```python
dataset.info()
```

```python
dataset['Total_Stops'].unique()
```

```python
dataset['Total_Stops'].mode()
```

```python
#'nan' because if you try to find out the mode of total_stop it will be 1
dataset['Total_Stops']=dataset['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.drop('Route',axis=1,inplace=True)
```
